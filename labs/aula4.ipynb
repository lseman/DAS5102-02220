```python
# Função que verifica se um array possui valores duplicados
def tem_valores_duplicados(array):
    for i in range(len(array)):
        for j in range(len(array)):
            if i != j and array[i] == array[j]:
                   return True
    return False
```

```python
# Função que verifica se um array possui valores duplicados usando enumerate
def tem_valores_duplicados(array):
    for i, valor_i in enumerate(array):
        for j, valor_j in enumerate(array):
            if i != j and valor_i == valor_j:
                return True
    return False
```

```python
def tem_valores_duplicados_rapida(arr):
    # Usando um conjunto para verificar duplicatas em tempo O(n)
    seen = set()
    for num in arr:
        if num in seen:
            return True  # Se já vimos o número, então há duplicata
        seen.add(num)
    return False  # Se nenhum duplicata for encontrada

# Teste da função tem_valores_duplicados_rapida
test_cases = [
    [64, 34, 25, 12, 22, 11, 90],
    [5, 2, 9, 1, 5, 6],
    [3, 0, 2, 5, -1, 4, 1],
    [7, 7, 7, 7, 7],
    [1, 2, 3, 4, 5]
]

results = [tem_valores_duplicados_rapida(arr) for arr in test_cases]
results
```

```python
# Função mais eficiente para verificar se há valores duplicados
def tem_valores_duplicados(array):
    return len(array) != len(set(array))
```

```python
def tem_valores_duplicados_rapida_print(arr):
    seen = set()
    step = 0
    for num in arr:
        step += 1
        print(f"Step {step}: Checking if {num} is in {seen}")
        if num in seen:
            print(f"Duplicate found: {num}")
            return True
        seen.add(num)
        print(f"Added {num} to {seen}")
    print("No duplicates found")
    return False

# Testando a função que imprime os passos
test_case = [1, 2, 3, 4, 2]
tem_valores_duplicados_rapida_print(test_case)
```
