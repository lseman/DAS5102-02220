```python
def selection_sort(arr):
    n = len(arr)
    # Itera sobre cada elemento da lista
    for i in range(n):
        # Encontra o índice do menor elemento no restante da lista
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Realiza a troca apenas se necessário
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Testando a função com diferentes listas de entrada
test_lists = [
    [64, 25, 12, 22, 11],
    [5, 4, 3, 2, 1],
    [1, 2, 3, 4, 5],
    [10, -1, 0, 5, 3],
    [9, 5, 7, 8, 2, 1, 6]
]

# Aplicando o selection_sort em cada lista
sorted_lists = [selection_sort(lst.copy()) for lst in test_lists]
sorted_lists
```

```python
import time

# Original selection sort implementation
def original_selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Optimized selection sort implementation
def optimized_selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        if min_idx != i:
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Test lists
test_lists = [
    [64, 25, 12, 22, 11],
    [5, 4, 3, 2, 1],
    [1, 2, 3, 4, 5],
    [10, -1, 0, 5, 3],
    [9, 5, 7, 8, 2, 1, 6],
    [i for i in range(1000, 0, -1)],  # Larger test case
    [i for i in range(1000)]           # Already sorted larger test case
]

# Measure time for original selection sort
start_time = time.time()
for lst in test_lists:
    original_selection_sort(lst.copy())
original_time = time.time() - start_time

# Measure time for optimized selection sort
start_time = time.time()
for lst in test_lists:
    optimized_selection_sort(lst.copy())
optimized_time = time.time() - start_time

original_time, optimized_time
```

```python
import time
import random
import pandas as pd

# Função para gerar o pior caso de seleção de trocas em todas as passagens
# Uma ideia é gerar uma lista em que os números maiores e menores alternem para sempre precisar de trocas
def generate_worst_case(size):
    # Alternando números grandes e pequenos
    return [i if i % 2 == 0 else size - i for i in range(size)]

# Função para medir o tempo de execução do selection_sort
def measure_execution_time(arr):
    start_time = time.time()
    selection_sort(arr)
    end_time = time.time()
    return end_time - start_time

# Testando em diferentes tamanhos de listas
sizes = [100, 500, 1000, 2000, 5000]
worst_case_times = []

# Medir o tempo de execução para cada tamanho de lista
for size in sizes:
    worst_case_list = generate_worst_case(size)
    exec_time = measure_execution_time(worst_case_list.copy())
    worst_case_times.append(exec_time)

# Exibindo o resultado
result_df = pd.DataFrame({
    "List Size": sizes,
    "Execution Time (s)": worst_case_times
})

result_df
```

```python
import matplotlib.pyplot as plt

# Implementação do bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # Últimos i elementos já estão no lugar
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

# Função para medir o tempo de execução do bubble sort
def measure_bubble_sort_time(arr):
    start_time = time.time()
    bubble_sort(arr)
    end_time = time.time()
    return end_time - start_time

# Medindo os tempos de execução do selection sort e bubble sort para diferentes tamanhos de listas
selection_sort_times = []
bubble_sort_times = []

for size in sizes:
    # Gerar a mesma lista para ambos os algoritmos
    worst_case_list = generate_worst_case(size)
    
    # Medir tempo do selection sort
    selection_time = measure_execution_time(worst_case_list.copy())
    selection_sort_times.append(selection_time)
    
    # Medir tempo do bubble sort
    bubble_time = measure_bubble_sort_time(worst_case_list.copy())
    bubble_sort_times.append(bubble_time)

# Plotar os tempos de execução comparando os algoritmos
plt.plot(sizes, selection_sort_times, label='Selection Sort', marker='o')
plt.plot(sizes, bubble_sort_times, label='Bubble Sort', marker='o')
plt.xlabel('Tamanho da lista')
plt.ylabel('Tempo de execução (segundos)')
plt.title('Comparação de Tempo de Execução: Selection Sort vs Bubble Sort')
plt.legend()
plt.grid(True)
plt.show()
```

# Características de um Passo
Um "passo" no código acima pode ser definido como uma iteração do loop for, onde a função verifica se o índice atual i é par (usando i % 2 == 0). Se verdadeiro, o elemento no índice i é adicionado a novo_array. Portanto, um passo inclui:

- A verificação de paridade do índice.
- A adição condicional de um elemento ao novo_array.

# Cenário de Pior Caso
No contexto desse problema, o pior caso ocorre quando a lista tem o maior tamanho possível para um dado valor n. Isso porque, independentemente dos valores dos elementos na lista, cada elemento é avaliado (a paridade do índice é verificada), o que significa que a função sempre iterará sobre todos os elementos da lista original. Assim, o tamanho da lista é o fator determinante do pior caso.

```python
def copia_um_sim_um_nao_print(array):
    novo_array = []
    for i in range(len(array)):
        print(f"Analisando índice {i} com valor {array[i]}")
        if i % 2 == 0:
            novo_array.append(array[i])
            print(f"Adicionando {array[i]} ao novo_array.")
    print(f"Array final: {novo_array}")
    return novo_array
```

# Complexidade da Função
A complexidade no tempo da função copia_um_sim_um_nao é O(n), onde n é o número de elementos no array de entrada. Isso porque a função precisa passar por todos os elementos da lista uma vez para verificar a condição de índice par e, potencialmente, copiar o elemento para o novo_array.

```python
import time

def copia_um_sim_um_nao(array):
    novo_array = []
    for i in range(len(array)):
        if i % 2 == 0:
            novo_array.append(array[i])
    return novo_array

# Gerar listas de tamanhos variados para teste
sizes = [1000, 5000, 10000, 50000, 100000]
execution_times = []

# Medir o tempo de execução para cada lista
for size in sizes:
    test_array = list(range(size))  # Gerar uma lista do tamanho especificado
    start_time = time.time()
    copia_um_sim_um_nao(test_array)
    end_time = time.time()
    execution_times.append(end_time - start_time)

# Plotar os tempos de execução
plt.figure(figsize=(10, 5))
plt.plot(sizes, execution_times, marker='o', linestyle='-', color='blue')
plt.title('Tempo de Execução da Função copia_um_sim_um_nao')
plt.xlabel('Tamanho da Lista')
plt.ylabel('Tempo de Execução (s)')
plt.grid(True)
plt.show()
```

```python
def insertion_sort(arr, left=0, right=None):
    if right is None:
        right = len(arr) - 1

    for i in range(left + 1, right + 1):
        key_item = arr[i]
        j = i - 1
        while j >= left and arr[j] > key_item:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key_item

    return arr

def merge(arr, left, mid, right):
    temp_arr = []
    i = left
    j = mid + 1

    while i <= mid and j <= right:
        if arr[i] <= arr[j]:
            temp_arr.append(arr[i])
            i += 1
        else:
            temp_arr.append(arr[j])
            j += 1

    while i <= mid:
        temp_arr.append(arr[i])
        i += 1
    while j <= right:
        temp_arr.append(arr[j])
        j += 1

    for i in range(len(temp_arr)):
        arr[left + i] = temp_arr[i]

def timsort(arr):
    min_run = 32
    n = len(arr)

    for start in range(0, n, min_run):
        end = min(start + min_run - 1, n - 1)
        insertion_sort(arr, start, end)

    size = min_run
    while size < n:
        for left in range(0, n, size * 2):
            mid = left + size - 1
            right = min((left + size * 2 - 1), (n - 1))

            if mid < right:
                merge(arr, left, mid, right)

        size *= 2

    return arr
```

```python
# Preparando o teste
sizes = [100, 500, 1000, 2000, 5000]  # Usaremos listas menores devido ao custo computacional
selection_sort_times = []
timsort_times = []

for size in sizes:
    test_array = [random.randint(0, 10000) for _ in range(size)]
    
    # Medindo o tempo para Selection Sort
    start_time = time.time()
    selection_sort(test_array.copy())
    selection_time = time.time() - start_time
    selection_sort_times.append(selection_time)
    
    # Medindo o tempo para Timsort
    start_time = time.time()
    timsort(test_array.copy())
    timsort_time = time.time() - start_time
    timsort_times.append(timsort_time)

# Plotar os resultados
plt.figure(figsize=(10, 5))
plt.plot(sizes, selection_sort_times, label='Selection Sort', marker='o')
plt.plot(sizes, timsort_times, label='Timsort', marker='o')
plt.xlabel('Tamanho da Lista')
plt.ylabel('Tempo de Execução (s)')
plt.title('Comparação de Tempo de Execução: Selection Sort vs Timsort')
plt.legend()
plt.grid(True)
plt.show()
```
