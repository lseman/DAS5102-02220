```python
def insertion_sort(arr):
    # Implementing the insertion sort algorithm
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        # Move elements that are greater than key to one position ahead of their current position
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Testing the function with different inputs
test_arrays = [
    [5, 2, 9, 1, 5, 6],
    [12, 11, 13, 5, 6],
    [2, 1, 0],
    [],
    [5],
    [-1, -3, -2, -5, -4],
]

# Applying insertion_sort to each test case
sorted_arrays = [insertion_sort(arr.copy()) for arr in test_arrays]
sorted_arrays
```

```python
import time
import matplotlib.pyplot as plt

# Function to generate worst-case lists (in descending order)
def generate_descending_list(size):
    return list(range(size, 0, -1))

# Function to measure the execution time of insertion_sort for different list sizes
def measure_insertion_sort_times():
    sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]  # Different list sizes
    times = []
    
    for size in sizes:
        test_list = generate_descending_list(size)
        start_time = time.time()  # Start the timer
        insertion_sort(test_list)  # Sort the list
        end_time = time.time()  # End the timer
        times.append(end_time - start_time)  # Record the execution time

    return sizes, times

# Measure and plot the times
sizes, times = measure_insertion_sort_times()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, marker='o')
plt.title('Execution Time of Insertion Sort (Worst Case) vs. List Size')
plt.xlabel('List Size')
plt.ylabel('Execution Time (seconds)')
plt.grid(True)
plt.show()
```

```python
# Defining the function 'interseccao' as described above
def interseccao(array1, array2):
    resultado = []
    for i in array1:
        for j in array2:
            if i == j:
                resultado.append(i)
    return resultado

# Test cases for the function 'interseccao'
test_cases = [
    ([1, 2, 3], [2, 3, 4]),
    ([], [1, 2, 3]),
    ([1, 2, 3], []),
    ([1, 2, 3, 4, 5], [5, 4, 3, 2, 1]),
    (['a', 'b', 'c'], ['c', 'd', 'e']),
    ([1, 2, 3, 3, 2, 1], [2])
]

# Applying 'interseccao' function to each test case
test_results = [interseccao(case[0], case[1]) for case in test_cases]
test_results
```

```python
def interseccao_rapida(array1, array2):
    # Converter um dos arrays em um conjunto para buscas rápidas
    set2 = set(array2)
    # Usar compreensão de lista para obter a interseção
    return [i for i in array1 if i in set2]

# Test cases
test_cases = [
    ([1, 2, 3], [2, 3, 4]),
    ([], [1, 2, 3]),
    ([1, 2, 3], []),
    ([1, 2, 3, 4, 5], [5, 4, 3, 2, 1]),
    (['a', 'b', 'c'], ['c', 'd', 'e']),
    ([1, 2, 3, 3, 2, 1], [2])
]

# Testando a função
for case in test_cases:
    print(f"Interseccao_rapida({case[0]}, {case[1]}) = {interseccao_rapida(case[0], case[1])}")
```

### Análise:

- **Complexidade de tempo**: A conversão de um array em um set ocorre em $O(m)$ e a iteração pelo outro array ocorre em $O(n)$. Assim, a complexidade da função é $O(n + m)$, que é muito mais eficiente do que $O(n \times m)$, especialmente para grandes conjuntos.
- **Eficiência**: A busca em um conjunto é muito mais rápida, $O(1)$, em comparação com uma busca linear, $O(n)$, o que reduz drasticamente o tempo de execução.

```python
import time
import matplotlib.pyplot as plt

# Alternative 1: Using dictionary
def interseccao_dicionario(array1, array2):
    dict2 = {i: True for i in array2}
    return [i for i in array1 if i in dict2]

# Alternative 2: Sort and Two-Pointer approach
def interseccao_two_pointer(array1, array2):
    array1.sort()
    array2.sort()
    i, j = 0, 0
    resultado = []
    while i < len(array1) and j < len(array2):
        if array1[i] == array2[j]:
            resultado.append(array1[i])
            i += 1
            j += 1
        elif array1[i] < array2[j]:
            i += 1
        else:
            j += 1
    return resultado

# Function to measure execution time of different methods
def measure_execution_time():
    sizes = [100, 200, 400, 800, 1600, 3200]  # Different list sizes
    times_original = []
    times_rapida = []
    times_dicionario = []
    times_two_pointer = []

    for size in sizes:
        array1 = list(range(size, 0, -1))
        array2 = list(range(size // 2, size + size // 2))

        # Measure original interseccao time
        start_time = time.time()
        interseccao(array1, array2)
        times_original.append(time.time() - start_time)

        # Measure interseccao_rapida time
        start_time = time.time()
        interseccao_rapida(array1, array2)
        times_rapida.append(time.time() - start_time)

        # Measure interseccao_dicionario time
        start_time = time.time()
        interseccao_dicionario(array1, array2)
        times_dicionario.append(time.time() - start_time)

        # Measure interseccao_two_pointer time
        start_time = time.time()
        interseccao_two_pointer(array1, array2)
        times_two_pointer.append(time.time() - start_time)

    return sizes, times_original, times_rapida, times_dicionario, times_two_pointer

# Measure and collect the execution times
sizes, times_original, times_rapida, times_dicionario, times_two_pointer = measure_execution_time()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times_original, marker='o', label='Original (O(n*m))')
plt.plot(sizes, times_rapida, marker='o', label='Rapida (O(n+m))')
plt.plot(sizes, times_dicionario, marker='o', label='Dicionario (O(n+m))')
plt.plot(sizes, times_two_pointer, marker='o', label='Two-Pointer (O(n log n))')
plt.title('Comparison of Execution Times for Different Intersection Methods')
plt.xlabel('List Size')
plt.ylabel('Execution Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
