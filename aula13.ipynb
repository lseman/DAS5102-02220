```python
import networkx as nx

# Criação do grafo direcionado vazio
G = nx.DiGraph()

# Adição de nós
G.add_node('Alice')
G.add_node('Bob')
G.add_node('Cynthia')

# Adição de arestas (ligando os nós)
G.add_edge('Alice', 'Bob')      # Aresta de Alice para Bob
G.add_edge('Bob', 'Cynthia')    # Aresta de Bob para Cynthia
G.add_edge('Cynthia', 'Alice')  # Aresta de Cynthia para Alice

# Exibe a lista de arestas
print("Arestas do grafo:", list(G.edges()))
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Criação do grafo
G = nx.DiGraph()

# Adição de nós e arestas
G.add_node('Alice')
G.add_node('Bob')
G.add_node('Cynthia')
G.add_edge('Alice', 'Bob')
G.add_edge('Bob', 'Cynthia')
G.add_edge('Cynthia', 'Alice')

# Definição de cores disponíveis
COLORS = ['#8dd3c7','#ffffb3','#bebada','#fb8072','#80b1d3','#fdb462',
          '#b3de69','#fccde5','#d9d9d9','#bc80bd','#ccebc5','#ffed6f']
# Experimentando diferentes parâmetros
nx.draw_circular(
    G, 
    node_color=COLORS[2],  # Cor dos nós: vermelho
    node_size=2500,  # Tamanho dos nós
    with_labels=True,  # Mostrar os rótulos dos nós
    font_weight='bold'  # Rótulos em negrito
)
plt.axis('equal')  # Ajusta os eixos para escala igual
plt.title("Grafo com Estilo Circular", fontsize=14)  # Adiciona título
plt.show()
```

```python
import networkx as nx

# Definição das posições dos nós
positions = {
    'Chicago': (-30, 30),
    'New York City': (30, 20),
    'Los Angeles': (-60, -50),
    'Miami': (25, -60)
}

# Criação do grafo não direcionado
G = nx.Graph()

# Adição dos nós a partir das chaves do dicionário positions
G.add_nodes_from(positions.keys())

# Verificação dos nós adicionados
nodes = list(G.nodes())
nodes
```

```python
# Definição das distâncias entre as cidades
distances = {
    ('Chicago', 'New York City'): 714,
    ('Chicago', 'Los Angeles'): 2014,
    ('Chicago', 'Miami'): 1381,
    ('New York City', 'Los Angeles'): 2451,
    ('Miami', 'Los Angeles'): 1400,
    ('New York City', 'Miami'): 1089
}

# Adição das arestas ao grafo
G.add_edges_from(distances.keys())

# Verificação dos nós e arestas adicionados
nodes = list(G.nodes())
edges = list(G.edges())

nodes, edges
```

```python
import matplotlib.pyplot as plt

# Desenho do grafo com a forma quadrada para os nós
nx.draw(
    G,
    pos=positions,  # Posições dos nós
    node_color='cyan',  # Cor dos nós
    node_size=2000,  # Tamanho dos nós
    with_labels=True,  # Exibir os rótulos dos nós
    node_shape='s',  # Forma dos nós: quadrado
    font_weight='bold'  # Texto em negrito
)

# Ajusta a escala dos eixos para melhor visualização
plt.axis('equal')
plt.title("Grafo com Nós em Formato de Quadrado", fontsize=14)
plt.show()
```

```python
# Desenho do grafo com etiquetas das arestas
plt.figure(figsize=(8, 6))

# Desenhando os nós e arestas
nx.draw(
    G,
    pos=positions,  # Posições dos nós
    node_color='cyan',  # Cor dos nós
    node_size=2000,  # Tamanho dos nós
    with_labels=True,  # Exibir os rótulos dos nós
    node_shape='s',  # Forma dos nós: quadrado
    font_weight='bold'  # Texto dos nós em negrito
)

# Etiquetas das arestas com os valores das distâncias
nx.draw_networkx_edge_labels(
    G,
    pos=positions,  # Posições dos nós
    edge_labels=distances,  # Dicionário com as etiquetas das arestas
    font_size=10  # Tamanho da fonte
)

plt.axis('equal')
plt.title("Grafo com Valores das Arestas", fontsize=14)
plt.show()
```
